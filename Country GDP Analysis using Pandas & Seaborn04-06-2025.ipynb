# we analyse python dataframe or dataset

```python
#introduction to seaborn # seaborn is very powerfull visualization(STATISTIC VISUALIZATION)pkg in python
import matplotlib.pyplot as plt # visualization
import seaborn as sns # distribution visualization
#seaborn are used for advance visualization e.x--> distribution plot, line plot

%matplotlib inline
plt.rcParams['figure.figsize'] = 6,2

import warnings
warnings.filterwarnings('ignore') #os error
```

```python
df.head()
```

```python
df['InternetUsers']
```

```python
# distributions:
vis1 = sns.distplot(df["InternetUsers"]) # UNIVRATE ANALYSIS - statistics
plt.show(vis1) # distplot = distribution plot graph
```

```python
vis2 = sns.displot(df["InternetUsers"])
plt.show(vis2)
```

```python
vis3 = sns.distplot(df["InternetUsers"], bins=15)
plt.show(vis3)
```

```python
vis4=sns.boxplot(data=df,x="IncomeGroup",y="BirthRate")
plt.show(vis4)
```

```python
vis5=sns.lmplot(data=df, x = 'InternetUsers', y = 'BirthRate')
plt.show(vis5)
```

```python
vis8 = sns.lmplot(data=df, x='InternetUsers',y='BirthRate',fit_reg=False, hue='IncomeGroup') # hue-- parameter for color
plt.show(vis8)
```
